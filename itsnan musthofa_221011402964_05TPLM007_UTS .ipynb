```python
#task 2

for company in company_detail_list:
    print(f"{company['name']}: {company['domain']} ({company['country']})")
```

```python
#task 3 

import json
company_domain_map = {company['name']: company['domain'] for company in company_detail_list}


employees_by_domain = {}
for employee in employee_detail_list:
    domain = company_domain_map[employee['company']]
    if domain not in employees_by_domain:
        employees_by_domain[domain] = []
    employees_by_domain[domain].append(employee['full_name'])

print(json.dumps(employees_by_domain, indent=4))
```

```python
#task 4


def get_employees_with_company_country():
    result = []
    
    # Membuat mapping antara nama perusahaan dan negara
    company_country_map = {company['name']: company['country'] for company in company_detail_list}
    
    for employee in employee_detail_list:
        full_name = employee['full_name']
        company = employee['company']
        country = company_country_map[company]
        
        result.append({
            "full_name": full_name,
            "company": company,
            "country": country
        })
    
    return result

# Mendapatkan hasil
employees_with_country = get_employees_with_company_country()

# Mencetak hasil dalam format yang diinginkan
import json  # Menggunakan json untuk format yang lebih baik
print(json.dumps(employees_with_country, indent=4))
```

```python
#task 5


def get_companies_with_employee_nationality():
   
    company_nationality = {}
    
    for employee in employee_detail_list:
        company = employee['company']
        nationality = employee['nationality']

        if company not in company_nationality:
            company_nationality[company] = {}

        if nationality not in company_nationality[company]:
            company_nationality[company][nationality] = 0
        
        company_nationality[company][nationality] += 1

   
    result = []
    for company, nationalities in company_nationality.items():
        result.append({
            "company": company,
            "employee_nationality": nationalities
        })
    
    return result


companies_with_nationality = get_companies_with_employee_nationality()


import json  # Menggunakan json untuk format yang lebih baik
print(json.dumps(companies_with_nationality, indent=4))
```
